### Analysis

- The results of 249 mice identified with SCC tumors were analyzed in order to evaluate the performance of the Capomulin treatment against other existing treatments.

- During the test, 5 replicate tests were found, which were excluded from the data set for the corresponding analyses. The sample was 50.4% male and 49.6% female. The evaluated set consisted of 10 drugs, among which a placebo was included.

- Capomulin and Ramicane showed more consistent results than the other drugs and as we can see in the box plot, Capomulin and Ramicane demonstrated better performance in controlling/reducing tumor size.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_s_c = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")

# Display the data table for preview
mouse_s_c.head()
```

```python
# Checking the number of mice.
n_mice = len(mouse_s_c["Mouse ID"].unique())

# Display the data table
n_mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
mouse_d = mouse_s_c.loc[mouse_s_c.duplicated(subset=['Mouse ID', 'Timepoint',]),'Mouse ID'].unique()
mouse_d
print (mouse_d)
```

```python
# Optional: Get all the data for the duplicate mouse ID.
mouse_ddf = mouse_s_c[mouse_s_c['Mouse ID'].isin(mouse_d)]

# Display the data table
mouse_ddf
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_notd = mouse_s_c[mouse_s_c["Mouse ID"].isin(mouse_d)==False]

# Display the data table
mouse_notd.head()
```

```python
# Checking the number of mice in the clean DataFrame.
mice_notd_n = len(mouse_notd["Mouse ID"].unique())

mice_notd_n
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
summary_sts = mouse_notd.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(['mean', 'median', 'var', 'std', 'sem'])

summary_sts = summary_sts.rename(columns={
    'mean': 'Mean Tumor Volume',
    'median': 'Median Tumor Volume',
    'var': 'Tumor Volume Variance',
    'std': 'Tumor Volume Std. Dev',
    'sem': 'Tumor Volume Std. Err.'})


# Assemble the resulting series into a single summary DataFrame.
summary_sts.head(20)
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
summary_sts_adv = mouse_notd.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(
    mean='mean',
    median ='median',
    var ='var',
    std='std',
    sem ='sem')



# Using the aggregation method, produce the same summary statistics in a single line
# Display the data table
summary_sts_adv.head(20)
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin
sel_regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
filtered_df = mouse_notd[mouse_notd["Drug Regimen"].isin(sel_regimens)]

# Start by getting the last (greatest) timepoint for each mouse
last_tests = filtered_df.groupby("Mouse ID")["Timepoint"].max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
last_tests = filtered_df.groupby("Mouse ID")["Timepoint"].max().reset_index()
final_volume_df = pd.merge(last_tests, filtered_df, on=["Mouse ID", "Timepoint"], how="left")

# Display the data table
final_volume_df.head(20)
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]


# Create empty list to fill with tumor vol data (for plotting)
vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers.
for treatment in treatments:

    # Locate the rows which contain mice on each drug and get the tumor volumes
    treatment_d = final_volume_df[final_volume_df["Drug Regimen"] == treatment]["Tumor Volume (mm3)"]

    # add subset
    vol_data.append(treatment_d)
    
    quartiles = treatment_d.quantile([.25, .5, .75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq

    # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)
    outliers = treatment_d[(treatment_d < lower_bound) | (treatment_d > upper_bound)]
    
    print(f"{treatment}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
fig1, ax1 = plt.subplots(figsize=(6, 4))
selected_regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
data_plot = [final_volume_df[final_volume_df["Drug Regimen"] == regimen]["Tumor Volume (mm3)"] for regimen in selected_regimens]
flierprops = dict(marker='o', markerfacecolor='red', markersize=8)
ax1.boxplot(data_plot, labels=selected_regimens, flierprops=flierprops)
ax1.set_ylabel('Final Tumor Volume (mm3)')

# Display the plot
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
capomulin_d = mouse_notd[mouse_notd["Drug Regimen"] == "Capomulin"]
mouse_id = "l509"
single_mouse_d = capomulin_d[capomulin_d["Mouse ID"] == mouse_id]
if not single_mouse_d.empty:
    plt.figure(figsize=(6, 5))
    plt.plot(single_mouse_d["Timepoint"], single_mouse_d["Tumor Volume (mm3)"], marker='o', linestyle='-', color='#31668a')
    
    plt.title(f'Campulin treatment of mouse l509')
    plt.xlabel('Time Point (days)')
    plt.ylabel('Tumor Volume (mm3)')

# Display the plot
    plt.show()

else:
    print(f"No data found for Mouse ID {mouse_id} treated with Capomulin.")
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
avg_volume = capomulin_d.groupby("Mouse ID").agg({
    "Tumor Volume (mm3)": "mean",
    "Weight (g)": "mean"  
}).reset_index()
plt.figure(figsize=(6, 5))
plt.scatter(avg_volume["Weight (g)"], avg_volume["Tumor Volume (mm3)"], color='#31668a', alpha=1.00)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

# Display the plot
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(avg_volume["Weight (g)"], avg_volume["Tumor Volume (mm3)"])
correlation_coefficient = correlation[0]
slope, intercept, r_value, p_value_regression, std_err = st.linregress(avg_volume["Weight (g)"], avg_volume["Tumor Volume (mm3)"])
x_values = avg_volume["Weight (g)"]
y_values = slope * x_values + intercept
plt.figure(figsize=(6, 5))
plt.scatter(avg_volume["Weight (g)"], avg_volume["Tumor Volume (mm3)"], color='#31668a', alpha=1.00, label='Observed Data')
plt.plot(x_values, y_values, color='red', linewidth=2, label='Linear Regression Line')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

# Display the plot
plt.show()
```
